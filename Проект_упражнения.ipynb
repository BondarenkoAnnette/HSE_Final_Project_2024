```python
#импорты
import pandas as pd

import re

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('universal_tagset')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords

import spacy
nlp = spacy.load("en_core_web_sm")
lemmatizer = nlp.get_pipe("lemmatizer")

import random
```

```python
#пробный текст
with open("/content/drive/MyDrive/sample_text.txt", "r", encoding='utf-8') as file:
  text = file.read()
```

```python
text
```

```python
#создадим датафрейм с токенами из текста
lst_tokens = []
lst_pos = []
lst_tags = []
lst_morph = []
lst_lemma = []

doc = nlp(text)

for token in doc:
    lst_tokens.append(token.text)
    lst_pos.append(token.pos_)
    lst_tags.append(token.tag_)
    lst_morph.append(token.morph)
    lst_lemma.append(token.lemma_)

df = pd.DataFrame({
            'Token': lst_tokens,
            'POS': lst_pos,
            'Tag': lst_tags,
            'Morphology': lst_morph,
            'Lemma': lst_lemma}
                )

df
```

```python
#создадим датафрейм только с глаголами и изучим их тэги
#для создания упражнения будем выбирать из текста только глаголы с тегами "VBZ", "VBN", "VBD", "VBG"
df_verbs = df[df['POS'] == 'VERB']
df_verbs
```

```python
from spacy import glossary
glossary.explain("VBN")
```

```python
#создаем функцию
def exercise(text):

    doc = nlp(text)
    tags = ["VBZ", "VBN", "VBD", "VBG"]
    verbs = []
    for token in doc:
      if (token.pos_ == "VERB") & (token.tag_ in tags):
        verbs.append(token)

    new_text = text
    for verb in verbs:
      reg_exp = fr"\s{verb}\s"
      reg_exp_2 = fr" [{verb.lemma_.upper()}] "
      new_text = re.sub(reg_exp, reg_exp_2, new_text)

    str = "Put the verbs in brackets in the correct form:"

    return(str, new_text)
```

```python
#проверим функцию на пробном тексте
exercise(text)
```
